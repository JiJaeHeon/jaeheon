분석 목적
“AI 개발자의 작업 성공 여부는 특정 요인들과 통계적으로 유의미한 관련이 있는가?”

```r
# 2단계: 데이터 취득과 정제

#  데이터 불러오기
data <- read.csv("/content/ai_dev_productivity.csv")
```

```r
# 구조 확인
str(data)
```

```r
# 결측치 확인
colSums(is.na(data))
```

구조 및 결측치 파악을 통해 누락된 값이 있는지 확인

이후 전처리의 기초 작업

결측치값이 0으로확인됨

```r
# 커피 1잔 = 약 150mg의 카페인 기준으로 변환
data <- data %>%
  mutate(coffee_cups = round(coffee_intake_mg / 150, 1))

# 예시 확인
data %>% select(coffee_intake_mg, coffee_cups) %>% head()
```

coffee_intake_mg는 단위가 명확하지 않아 해석이 어렵기 때문에, 일상적으로 이해 가능한 단위인 잔 수로 환산합니다.

새로운 변수 coffee_cups 생성

```r
# 인지부담 점수 범위 정의
# 0~3: 낮음, 4~6: 중간, 7~10: 높음
data <- data %>%
  mutate(cognitive_category = case_when(
    cognitive_load <= 3 ~ "Low",
    cognitive_load <= 6 ~ "Medium",
    TRUE ~ "High"
  ))

# factor로 변환 (모델링 편의 위해)
data$cognitive_category <- factor(data$cognitive_category, levels = c("Low", "Medium", "High"))

# 결과 예시 확인
data %>% select(cognitive_load, cognitive_category) %>% head()
```

연속적인 수치(0~10)를 세분화하여 해석 가능하게 함

Low (0~3) → 부담 거의 없음

Medium (4~6) → 다소 집중 필요

High (7~10) → 과부하 상태

추후 시각화나 분류 모델에 사용 시 해석 용이성 증가

```r
# task_success는 분류용 변수이므로 factor로 변환
data$task_success <- as.factor(data$task_success)
```

task_success는 0/1로 이진 분류 모델의 타깃이기 때문에 factor로 바꿔야 R의 분류 모델들이 올바르게 인식함

```r
# 1. 수치형 변수 표준화 (z-score)
# 대상: hours_coding, sleep_hours, commits, ai_usage_hours, cognitive_load, coffee_cups
data_scaled <- data %>%
  mutate(across(c(hours_coding, sleep_hours, commits, ai_usage_hours, cognitive_load, coffee_cups),
                ~ scale(.)[,1],
                .names = "scaled_{.col}"))

# 2. 파생 변수 제거/유지 정리
# - coffee_intake_mg 제거 (coffee_cups로 대체)
# - cognitive_load 원본은 범주화했으므로 분석에서는 제외
data_clean <- data_scaled %>%
  select(-coffee_intake_mg, -cognitive_load)

# 최종 확인
str(data_clean)
```

scale() 함수는 평균 0, 표준편차 1로 정규화해줍니다.

이 작업은 SVM, k-NN, 로지스틱 회귀 등 거리 기반 모델에서 필수적입니다.

변수 간 단위 차이(ex. 시간 vs 점수)가 클 경우 모델 성능이 악화될 수 있기 때문입니다.

coffee_intake_mg는 이제 coffee_cups로 대체했기 때문에 제거

cognitive_load는 원본 수치를 범주로 나누어 cognitive_category를 사용하므로 제거

hours_coding, sleep_hours, commits, ai_usage_hours, cognitive_load 이변수들만 z-스코어를 적용한 이유 값들이 연속형 숫자이기때문에

distractions, bugs_reported 이친구들은 이산형 카운트 변수이기 떄문에 제외

| 구분       | 변수명                                                                                                             |
| -------- | --------------------------------------------------------------------------------------------------------------- |
| ✔ 원본 유지  | `hours_coding`, `distractions`, `sleep_hours`, `commits`, `bugs_reported`, `ai_usage_hours`, `task_success`     |
| ➕ 파생 추가  | `coffee_cups`, `cognitive_category`                                                                             |
| ➕ 표준화 추가 | `scaled_hours_coding`, `scaled_sleep_hours`, `scaled_commits`, `scaled_ai_usage_hours`, `scaled_cognitive_load` |


```r
# 모든 독립 변수와 종속 변수 task_success(0/1) 간의 관계를 시각적으로 분석하여
# 어떤 변수가 성공/실패에 영향을 줄 가능성이 높은지 파악합니다.

#  1. 수치형 변수 리스트
box_vars <- c("hours_coding", "sleep_hours", "commits", "ai_usage_hours",
              "coffee_cups", "distractions", "bugs_reported")

#  2. 수치형 변수별 Boxplot을 반복 출력
for (var in box_vars) {
  print(
    ggplot(data_clean, aes_string(x = "task_success", y = var, fill = "task_success")) +
      geom_boxplot() +
      labs(
        title = paste("작업 성공 여부에 따른", var, "분포"),
        x = "작업 성공 여부",
        y = var
      ) +
      theme_minimal()
  )
}

#  3. 범주형 변수: cognitive_category vs task_success 비율 barplot
print(
  ggplot(data_clean, aes(x = cognitive_category, fill = task_success)) +
    geom_bar(position = "fill") +
    scale_y_continuous(labels = scales::percent) +
    labs(
      title = "인지부담 범주별 작업 성공률",
      x = "인지부담",
      y = "성공률 비율"
    ) +
    theme_minimal()
)
```

가설 설정
귀무가설 (H₀): 설명 변수들은 작업 성공 여부(task_success)에 통계적으로 유의미한 영향을 미치지 않는다.

대립가설 (H₁): 설명 변수들 중 하나 이상은 작업 성공 여부(task_success)에 통계적으로 유의미한 영향을 미친다.

```r
# 전에한 정제는“탐색/시각화용”이었고,
# 지금 하는 정제는 “모델 학습에 맞게 데이터 포맷을 따로 구성”하는 것

# 1. 변수 선택 (scaled 변수 + 정수형 + 범주형만)
data_model <- data_clean %>%
  select(
    task_success,
    scaled_hours_coding,
    scaled_sleep_hours,
    scaled_commits,
    scaled_ai_usage_hours,
    scaled_coffee_cups,
    distractions,
    bugs_reported,
    cognitive_category
  )

# 2. dummyVars 적용 (target은 제외하고 더미 처리)
dummy <- dummyVars(~ ., data = data_model %>% select(-task_success))
data_dummy <- predict(dummy, newdata = data_model %>% select(-task_success)) %>%
  as.data.frame()

# 3. target 다시 붙이기
data_ready <- cbind(data_dummy, task_success = data_model$task_success)

# 4. train/test 분할
set.seed(123)
train_index <- createDataPartition(data_ready$task_success, p = 0.8, list = FALSE)
train <- data_ready[train_index, ]
test <- data_ready[-train_index, ]
```

```r
# 모델 학습
model_logit <- glm(task_success ~ ., data = train, family = "binomial")

# 예측 확률
pred_logit <- predict(model_logit, newdata = test, type = "response")
pred_logit_class <- ifelse(pred_logit > 0.5, 1, 0)

# 평가
confusionMatrix(factor(pred_logit_class), test$task_success)

# 회귀 계수 확인
summary(model_logit)
```

```r
# 회귀 계수 시각화 (로지스틱 회귀)
coeffs <- summary(model_logit)$coefficients
coeffs_df <- data.frame(
  Variable = rownames(coeffs),
  Estimate = coeffs[, "Estimate"]
)

# 불필요한 항 제거 (Intercept, NA)
coeffs_df <- coeffs_df %>% filter(!is.na(Estimate) & Variable != "(Intercept)")

# 시각화
ggplot(coeffs_df, aes(x = reorder(Variable, Estimate), y = Estimate)) +
  geom_col(fill = "steelblue") +
  coord_flip() +
  labs(title = "로지스틱 회귀 계수 시각화", x = "변수", y = "회귀 계수(Estimate)") +
  theme_minimal()
```

```r

# 확률 예측값 이미 구한 상태: pred_logit
roc_logit <- roc(test$task_success, pred_logit)

# ROC 곡선 그리기
plot(roc_logit, col = "blue", main = "로지스틱 회귀 - ROC 곡선")
auc(roc_logit)
```

[2] ROC 곡선 (모델 성능 평가)
ROC 곡선은 모델이 **양성 클래스(작업 실패)**를 얼마나 잘 분류하는지 나타냅니다.

X축 (1 - Specificity): 거짓 양성률

Y축 (Sensitivity): 진짜 양성률

▶ 해석:
파란선은 모델의 예측 성능입니다.

곡선이 왼쪽 위로 굽어질수록 좋은 모델입니다.

**ROC 곡선 아래 면적 (AUC, Area Under Curve)**가 0.9 이상일 가능성이 높아 보이며, 이는 매우 우수한 성능을 의미합니다.

곡선이 왼쪽 위에 가까울수록 성능이 우수

AUC 1에 가까우면 이상적 (보통 0.8 이상이면 훌륭)

종합 해석
**정확도(Accuracy): 약 87%**로 우수한 분류 성능을 보이고 있습니다.

**중요 변수들(커피, 수면, 코딩 시간, 인지부담 중간 등)**이 성공에 긍정적인 영향을 미친다는 통계적 증거가 있습니다.

모델은 실제 실패와 성공을 잘 분류해내며, 특히 **특이도(90%)**와 민감도(82%) 모두 균형 있게 높습니다.

```r
# 로지스틱 회귀 모델 학습
glm_model <- glm(task_success ~ ., data = train, family = "binomial")

# 로지스틱 회귀 계수 추출
coefs <- summary(glm_model)$coefficients[-1, 1]  # Intercept 제외

# 데이터프레임 변환
coef_df <- data.frame(
  Feature = names(coefs),
  Coefficient = coefs
)

# 히트맵
library(ggplot2)
ggplot(coef_df, aes(x = "", y = reorder(Feature, Coefficient), fill = Coefficient)) +
  geom_tile() +
  scale_fill_gradient2(low = "red", mid = "white", high = "blue", midpoint = 0) +
  labs(title = "로지스틱 회귀 계수 히트맵", x = "", y = "Feature") +
  theme_minimal()
```

① 로지스틱 회귀 계수 히트맵

원형: 일반 샘플

색깔: 실제 정답 값

해석:

SVM은 결정 경계를 만드는 데 꼭 필요한 경계선 근처 샘플만 이용합니다.

이 삼각형 모양들이 바로 서포트 벡터입니다.

성공과 실패 사이에 어떤 샘플들이 경계를 결정짓는 데 중요한지 시각적으로 확인할 수 있습니다.

이 시각화는 모델 복잡성 이해와 오버피팅 여부 판단에도 유용합니다.

```r

# 확률 기반 예측과 실제값 차이 계산
svm_diff <- abs(svm_pred_prob - as.numeric(test$task_success))

# 상관 기반 영향도 계산
svm_heat <- cor(test %>% select(-task_success), svm_diff)

# 행렬 형태로 변환
heat_df <- data.frame(Feature = rownames(svm_heat), Influence = as.vector(svm_heat))

# 히트맵
ggplot(heat_df, aes(x = "", y = reorder(Feature, Influence), fill = Influence)) +
  geom_tile() +
  scale_fill_gradient(low = "white", high = "darkred") +
  labs(title = "SVM 예측 오차 기반 변수 영향 히트맵", x = "", y = "Feature") +
  theme_minimal()
```

```r

# 모델 학습 (다시 실행)
dt_model <- rpart(task_success ~ ., data = train_clean, method = "class")

# 클래스 예측
dt_pred_class <- predict(dt_model, newdata = test_clean, type = "class")

# 확률 예측 (AUC 계산용)
dt_pred_prob <- predict(dt_model, newdata = test_clean, type = "prob")[, "1"]
```

```r
#모델성능비교

# 성능 수치 저장
model_results <- data.frame(
  Model = c("로지스틱 회귀", "결정 트리", "SVM", "랜덤 포레스트"),
  Accuracy = c(
    mean(pred_logit_class == test$task_success),
    mean(dt_pred_class == test_clean$task_success),
    mean(svm_pred_class == test_clean$task_success),
    mean(rf_pred_class == test_clean$task_success)
  ),
  AUC = c(
    auc(roc(test$task_success, pred_logit)),
    auc(roc(test_clean$task_success, dt_pred_prob)),
    auc(roc(test_clean$task_success, svm_pred_prob)),
    auc(roc(test_clean$task_success, rf_pred_prob))
  )
)

# 소수점 반올림
model_results <- model_results %>%
  mutate(across(c(Accuracy, AUC), round, 4))

# 출력
print(model_results)
```

1. Accuracy 해석 (정확도)

랜덤 포레스트 모델이 가장 높은 정확도를 보여주며, 실제 task_success (업무 성공 여부)를 잘 예측합니다.

로지스틱 회귀는 성능이 제일 낮지만 여전히 86.9%로 나쁘지 않습니다.

2. AUC 해석 (Area Under the ROC Curve)
AUC는 모델이 클래스를 구분하는 능력을 측정합니다.
값이 1에 가까울수록 이진 분류 구분을 잘함을 의미합니다.

랜덤 포레스트: AUC = 0.9944 → 거의 완벽한 분류

SVM: AUC = 0.9825 → 아주 강력한 구분 능력

로지스틱 회귀: AUC = 0.9397, 결정 트리: AUC = 0.9321 → 상당히 우수하지만 상대적으로 낮음

```r
# 성능 데이터프레임 생성
model_compare <- data.frame(
  Model = c("로지스틱 회귀", "결정 트리", "SVM", "랜덤 포레스트"),
  Accuracy = c(0.8687, 0.9293, 0.9091, 0.9394),
  AUC = c(0.9397, 0.9321, 0.9825, 0.9944)
)

# 데이터 가공 (long format for ggplot2)
model_long <- pivot_longer(model_compare, cols = c(Accuracy, AUC), names_to = "Metric", values_to = "Score")

# 시각화
ggplot(model_long, aes(x = Model, y = Score, fill = Metric)) +
  geom_bar(stat = "identity", position = "dodge") +
  labs(title = "모델 성능 비교 (정확도 & AUC)", y = "Score", x = "모델") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 15, hjust = 1)) +
  scale_fill_manual(values = c("skyblue", "coral"))
```
